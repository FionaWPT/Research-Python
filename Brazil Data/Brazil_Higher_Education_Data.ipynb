```python
# import the library
import pandas as pd
import numpy as np
import googletrans
from deep_translator import GoogleTranslator
```

## Importing the datasets

```python
# Main dataset we are using to get the university name
ies=pd.read_csv('ies.csv')
ies.head()
```

```python
# second dataset we are using to get the type and area of the universities
cpc=pd.read_csv('cpc.csv')
cpc.head()
```

## Cleaning the Two Datasets

```python
## getting 2018 data only from ies datasets
ies_2018=ies[ies['NU_ANO']==2018]
## keep the NO_IES
ies_2018=ies_2018['NO_IES'].reset_index().drop(columns=['index'])
ies_2018.columns=['college']
ies_2018
```

```python
len(ies_2018['college'].unique()) #there is duplicates in college
```

```python
ies_2018=ies_2018.drop_duplicates()
ies_2018
```

```python
# Cleaning cpc with only Nome da IES(name),Categoria Administrativa(type), Município do Curso(area) from 2018
cpc_2018=cpc[cpc['Ano']==2018]
cpc_2018=cpc_2018[['Nome da IES','Categoria Administrativa','Município do Curso']].reset_index().drop(columns=['index'])
cpc_2018.columns=['college','unitype','city']
cpc_2018=cpc_2018.drop_duplicates()
cpc_2018

##len(cpc_2018['college'].unique())
```

## Merging two datasets to one

```python
ies_cpc=pd.merge(ies_2018,cpc_2018,how='left')
ies_cpc
```

```python
#missing data in each column
ies_cpc.isna().sum() 
```

## Filling out some of the missing data in unitype column

```python
#some colleges that have missing type in ies_cpc can be found in the dataset igc
igc=pd.read_csv('igc.csv')
igc_2018=igc[igc['Ano']==2018]
igc_2018=igc[igc['NM_IES']=='FACULDADE DE MEDICINA DE ITAJUBÁ']
igc_2018=igc[['NM_IES','NM_CATEGORIA_ADMINISTRATIVA']]
igc_2018.columns=['college','unitype']
igc_2018=igc_2018.drop_duplicates(subset='college',keep='first')
igc_2018
```

```python
#all the rows that have null unitype
null_data=ies_cpc[ies_cpc['unitype'].isnull()]
null_data
```

```python
#filling out some of the missing unitype
for uni in null_data['college']:
    for other in igc_2018['college']:
        if uni==other:
            ind=ies_cpc[ies_cpc['college']==uni].index
            ies_cpc.loc[ind,'unitype']=igc_2018.loc[igc_2018['college']==uni,'unitype'].iloc[0]
ies_cpc
```

```python
#missing unitype in ies_cpc changed from 792 to 435
ies_cpc[ies_cpc['unitype'].isnull()]
```

## Uniform text in unitype column

```python
#check how many different entries are in unitype
ies_cpc['unitype'].unique()
```

```python
#a dataframe with all the non-null unitype
no_null=ies_cpc[ies_cpc['unitype'].notnull()]
```

```python
#uniform the unitype column by private or public
for i in no_null.index:
    unitype=ies_cpc.loc[i,'unitype'].lower()
    if 'pública' in unitype or 'público' in unitype:
        ies_cpc.loc[i,'unitype']='Public'
    elif 'privado' in unitype or 'privada' in unitype:
        ies_cpc.loc[i,'unitype']='Private'
    else:
        ies_cpc.loc[i,'unitype']=ies_cpc.loc[i,'unitype']
ies_cpc
```

```python
#see how many different entries are in unitype column after cleaning
ies_cpc['unitype'].unique()
```
